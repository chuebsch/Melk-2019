```python
%matplotlib inline

import pickle
import argparse
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
from sklearn.externals import joblib
from sklearn.preprocessing import StandardScaler
```

# Import classifier model

```python
with open('best_tree_rand_20190815_0845.pkl', 'rb') as f:
  model = joblib.load(f)
```

# Import data to predict

```python
predict = pd.read_csv("predict.csv", na_filter=False, skipinitialspace=True, thousands=',')
print(predict.head())

to_predict = predict[['wavelength', 'mr_reso', 'mr_sg_no', 'eLLG', 'tncs',
       'model_res', 'anomalousCC', 'IoverSigma', 'completeness', 'diffI',
       'RmergeI', 'lowreslimit', 'RpimI', 'multiplicity', 'RmeasdiffI',
       'anomalousslope', 'diffF', 'wilsonbfactor', 'RmeasI', 'highreslimit',
       'RpimdiffI', 'anomalousmulti', 'RmergediffI', 'totalobservations',
       'anomalouscompl', 'cchalf', 'totalunique']]

print(to_predict.shape)
```

# Fill in missing values

```python
to_predict_fill = to_predict.fillna(0)
```

# Predict

```python
unknown = to_predict.values

for line in unknown:
  y_pred = model.predict(line.reshape(1, -1))
  y_pred_proba = model.predict_proba(line.reshape(1, -1))
  fail_prob = round(y_pred_proba[0][0], 4) * 100
  succ_prob = round(y_pred_proba[0][1], 4) * 100
  y_pred_adj = [1 if x >= 0.9317 else 0 for x in y_pred_proba[:, 1]]
  print('Probability for molecular replacement outcome:')
  print('Failure: %s' %fail_prob)
  print('Success: %s' %succ_prob)
  print('*' * 80)

#make predictions
ynew = model.predict(unknown)
# show the inputs and predicted outputs
for i in range(len(unknown)):
  print("X=%s, Predicted=%s" % (unknown[i], ynew[i]))
  
# make probabilistic predictions
ynew_proba = model.predict_proba(unknown)
# show the inputs and predicted probabilities
for i in range(len(unknown)):
  print("X=%s, Predicted=%s" % (unknown[i], ynew_proba[i]))
```
